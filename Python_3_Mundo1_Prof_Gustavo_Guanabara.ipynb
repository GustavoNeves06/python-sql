```python
# Exercício 11 - Aula 7

largura = float(input('Qual a largura da parede em metros: '))
altura = float(input('Qual a altura da parede em metros: '))
area = largura * altura
litro = area / 2

print('A área dessa parede é de {:.2f} metros quarados, e vou precisar de {:.2f} litros de tinta.'.format(area, litro))
```
