```python
def double_number(number):
  return number * 2

result = double_number(5)
print(result)
```

```python
def triple_number(number):
  return number * 3

result = triple_number(8)
print(result)
```

```python
def check_even_odd(number):
  if number % 2 == 0:
    return 'Even'
  else:
    return 'Odd'

print(check_even_odd(10))
print(check_even_odd(7))

```

```python
def validate_meetings(contacts, meetings):
  if contacts >= meetings:
    return 'Valid'
  else:
    return 'Invalid'

print(validate_meetings(100, 30))
print(validate_meetings(50, 70))
```
