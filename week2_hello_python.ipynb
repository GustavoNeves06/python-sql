```python
def validate_funnel(contacts, meetings, closed_deals):
  if meetings > contacts:
    return 'Invalid: meetings greater than contacts'
  elif closed_deals > meetings:
    return 'Invalid: deals greater than meetings'
  elif contacts < 0 or meetings < 0 or closed_deals < 0:
    return 'Invalid: negative values'
  else:
    return 'Valid'

print(validate_funnel(100, 30, 10))   # Valid
print(validate_funnel(50, 70, 10))    # meetings > contacts
print(validate_funnel(100, 30, 40))   # deals > meetings
print(validate_funnel(-10, 5, 2))     # negative values
```

```python
def validate_funnel(contacts, meetings, closed_deals):
  if meetings > contacts:
    return 'Invalid: meetings greater than contacts'
  elif closed_deals > meetings:
    return 'Invalid: deals greater than meetings'
  elif contacts < 0 or meetings < 0 or closed_deals < 0:
    return 'Invalid: negative values'
  else:
    return 'Valid'

def analyze_funnel(contacts, meetings, closed_deals):

    validation = validate_funnel(contacts, meetings, closed_deals)

    if validation != 'Valid':
        return validation

    conversion_rate = (meetings / contacts) * 100
    close_rate = (closed_deals / meetings) * 100

    if close_rate >= 15:
        health = 'Good'
    else:
        health = 'Bad'

    return {
        'conversion_rate': conversion_rate,
        'close_rate': close_rate,
        'health': health
    }


print(analyze_funnel(200, 40, 8))
print(analyze_funnel(50, 70, 10))
```
